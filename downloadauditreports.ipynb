```python
import os
import pandas as pd
import requests

df = pd.read_excel('/Users/lojainabdulla/Downloads/Sources for dataset/Finaldataset.xlsx')

def download_reports(df, url_col, year_col, label_col, folder):
    os.makedirs(folder, exist_ok=True)
    for _, row in df.iterrows():
        url, year, label = row[url_col], row[year_col], str(row[label_col]).replace('/', '-').replace('\\', '-')
        if pd.isna(url) or not url.strip(): continue
        print(f"Checking row: Year={year}, Label={label}, URL={url}")
        try:
            r = requests.get(url, timeout=20)
            r.raise_for_status()
            with open(os.path.join(folder, f"{year}_{label}.pdf"), 'wb') as f:
                f.write(r.content)
            print(f"Downloaded: {year}_{label}.pdf")
        except Exception as e:
            print(f"Failed: {url} ({e})")

download_reports(df, "AuditorReportOrLetterLink", "Year", "Row Labels", "downloaded_reports")
```
